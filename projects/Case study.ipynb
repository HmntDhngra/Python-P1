### Problem 1: Find mean, median, IQR for Sachin, Rahul, and India

```python
heads=[Sachin, Rahul, India]
heads_name = ['Sachin', ' Rahul', 'India']
type(heads)
#Made a list of people
```

```python
for i,j in zip(heads, heads_name):
    i_mean=np.mean(i)
    i_median=np.median(i)
    i_IQR=np.percentile(i,75)-np.percentile(i,25)
    print(j,"Mean", i_mean)
    print(j,"Median", i_median)
    print(j,"IQR", i_IQR)
    print("")
```

```python
def stats(i):
    print("mean", np.mean(i))
    print("median", np.median(i))
    print("IQR", np.percentile(i,75)-np.percentile(i,25))
```

```python
stats(Sachin)
```

```python
stats(Rahul)
```
